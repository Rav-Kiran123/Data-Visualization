```python
year=[2010,2011,2012,2013,2014,2015]
yield_apples=[0.895,0.91,0.919,0.926,0.929,0.931]
plt.plot(year,yield_apples)
```

```python
plt.plot(year,yield_apples)
plt.xlabel('year')
plt.ylabel('Yield of Apples')
```

```python
year=[2010,2011,2012,2013,2014,2015]
yield_apples=[0.895,0.91,0.919,0.926,0.929,0.931]
yield_orange=[0.962,0.941,0.923,0.928,0.956,0.962]
plt.plot(year,yield_apples)
plt.plot(year,yield_orange)
plt.xlabel('years')
plt.ylabel('Yield of apples and Orange')
```

```python
plt.plot(year,yield_apples, marker='o')
plt.plot(year,yield_orange, marker='x')
plt.xlabel('years')
plt.ylabel('Yield of apples and Orange')
```

```python
plt.plot(year,yield_apples, 's-b')
plt.plot(year,yield_orange, 'o--r')
plt.xlabel('years')
plt.ylabel('Yield of apples and Orange')
```

```python
# Distribution Plot
```

```python
sns.distplot(data.total_bill)
```

```python
# Scatter Plot
sns.scatterplot(data.total_bill,data.tip)
```

```python
# Joint Plot
sns.jointplot(data.total_bill,data.tip)
```

```python
# Boxplot
sns.boxplot(data.total_bill)
```

```python
#bar plot
sns.barplot(data.total_bill,data.tip)
```

```python
sns.barplot('day','total_bill',data=data)
```

```python
data.corr() #-- It is used to find the correlation among the columns in DataFrame
```

```python
#Heat Map
#Graphical Representation of Dara using Colors to visualise the data in Matrix
# Common values will be brighter in color
# Less common values will be darker in color
hm=data.corr()
sns.heatmap(hm)
```
